```python
import datetime as dt

import numpy as np
import pandas as pd

from plotly import tools
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

```python
df = pd.read_csv('data.csv')
df['CrimeDate'] = pd.to_datetime(df['CrimeDate'])
```

```python
pd.Series.unique(df['Description'])
```

```python
burg = df[df['Description'] == 'BURGLARY']
shoot = df[df['Description'] == 'LARCENY']
arson = df[df['Description'] == 'ARSON']
```

```python
burg = burg.groupby('CrimeDate').CrimeDate.count().reset_index(name="count")
shoot = shoot.groupby('CrimeDate').CrimeDate.count().reset_index(name="count")
arson = arson.groupby('CrimeDate').CrimeDate.count().reset_index(name="count")
```

```python
burg_april = burg.loc[(burg['CrimeDate'] > '2015-4-1') & (burg['CrimeDate'] <= '2015-5-15')]
shoot_april = shoot.loc[(shoot['CrimeDate'] > '2015-4-1') & (shoot['CrimeDate'] <= '2015-5-15')]
arson_april = arson.loc[(arson['CrimeDate'] > '2015-4-1') & (arson['CrimeDate'] <= '2015-5-15')]
```

```python
fig = tools.make_subplots(rows=1, cols=2, subplot_titles=(
    '2012-2017', 'April - May 2015'
))

trace0 = go.Scatter(x=burg['CrimeDate'], y=burg['count'], name='Burglaries')
fig.append_trace(trace0, 1, 1)

trace1 = go.Scatter(x=shoot['CrimeDate'], y=shoot['count'], name='Larceny')
fig.append_trace(trace1, 1, 1)

trace2 = go.Scatter(x=arson['CrimeDate'], y=arson['count'], name='Arson')
fig.append_trace(trace2, 1, 1)

trace3 = go.Scatter(x=burg_april['CrimeDate'], y=burg_april['count'], name='Burglaries')
fig.append_trace(trace3, 1, 2)

trace4 = go.Scatter(x=shoot_april['CrimeDate'], y=shoot_april['count'], name='Larceny')
fig.append_trace(trace4, 1, 2)

trace5 = go.Scatter(x=arson_april['CrimeDate'], y=arson_april['count'], name='Arson')
fig.append_trace(trace5, 1, 2)

fig['layout'].update(title='City in Protest: Baltimore Crimes during April 2015 Riots')
fig['layout'].update(showlegend=False)
fig['layout']['yaxis1'].update(title='Reported Incidents')
fig['layout']['yaxis2'].update(title='Reported Incidents')
fig['layout']['xaxis1'].update(nticks=6)
fig['layout']['xaxis2'].update(nticks=6)

py.iplot(fig, filename='baltimore crimes')
```
